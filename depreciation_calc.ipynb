```python
import pandas as pd
pd.options.display.float_format = '{:.0f}'.format

print("WELCOME TO DEPRECIATION CALCULATOR!!!")
print("\n")

print ("Plese choose method which you use to depreciate your PP&E")
print("\n")
print (" SL method : Straight Line Method", "\n", "SYD method : Sum of Years Depreciation Method", "\n", "DDB method : Double Decline Balance Method", "\n")
x = input("Fill 1 for SL method, 2 for SYD method, and 3 for DDB method ")
x = int(x)
print("\n")

print("In the sections below, you will fill Year of Purchase, End Year, Acqution cost, Salvage Value, and Useful Life Period(this is optional)")
print("\n")

print("Warning!", "\n", "-If you fill in year of purchase and end year section, please fill 'No' in useful life period section.", "\n", "-If you fill in useful life period section, please fill 'No' in both year of purchase and end year section.", "\n", "-If you choose DDB method, make sure to fill '0' in Salvage Value section.", "\n","-(The scenario is that you buy a PP&E on JAN 1st of the year of purchase and depreciation will end on DEC 31st the end year.")
print("\n")

bYear = input("Year of Purchase")
if bYear == "No":
  bYear = None
else:
  bYear = int(bYear)
print("\n")
eYear = input("End Year")
if eYear == "No":
  eYear = None
else:
  eYear = int(eYear)
print("\n")
aCost = input("Acquition Cost")
aCost = int(aCost)
print("\n")
sValue = input("Salvage Value")
sValue = int(sValue)
print("\n")
period = input("(Useful Life Period)")
if period == "No":
  period = None
else:
  period = int(period)
print("\n")



def SLmethod (bYear, eYear, aCost, sValue, period):
  if period == None:
    period = eYear - bYear + 1
  else:
    bYear = 0

  
  yearCount = []
  dList = [] 
  adList = [] 
  svList = []

  for i in range (period):
    yearCount.append(bYear + i)
    dList.append((aCost-sValue)/period)
    adList.append(sum(dList))
    svList.append(aCost - adList[i])
  
  resultList = [dList, adList, svList]

  df1 = pd.DataFrame(resultList, columns = yearCount , index = ["DE", "AD", "SV"])

  
  print("'DE' = Depreciation amount of each year" + "\n" + "'AD' = Accumulated depreciation amount" + "\n" + "'SV' = Salvage Value")
  print("\n" + "~Result~")
  print("1)On Jan 1st", bYear, ": AD = 0", "SV =", aCost, "\n")
  print("2)The end of term of each year")
  print(df1.T)

def SYDmethod(bYear, eYear, aCost, sValue, period):
  if period == None:
    period = eYear - bYear + 1
  else:
    bYear = 0

  yearCount = []
  periodCount = []
  dList = []  
  adList = [] 
  svList = [] 

  for i in range (period):
    periodCount.append(period - i)

  for i in range (period):
    yearCount.append(bYear + i)
    dList.append((aCost-sValue) * (periodCount[i] / sum(periodCount)))
    adList.append(sum(dList))
    svList.append(aCost - adList[i])
  
  resultList = [dList, adList, svList]

  df1 = pd.DataFrame(resultList, columns = yearCount , index = ["DE", "AD", "SV"])

  
  print("'DE' = Depreciation amount of each year" + "\n" + "'AD' = Accumulated depreciation amount" + "\n" + "'SV' = Salvage Value")
  print("\n" + "~Result~")
  print("1)On Jan 1st", bYear, ": AD = 0", "SV =", aCost, "\n")
  print("2)The end of term of each year")
  print(df1.T)

def DDBmethod (bYear, eYear, aCost,sValue, period):
  if period == None:
    period = eYear - bYear + 1
  else:
    bYear = 0
  
  yearCount = []
  dList = [] 
  adList = [0]
  svList = [] 

  for i in range (period):
    yearCount.append(bYear + i)
    dList.append((aCost - adList[i]) * (2/period))
    adList.append(sum(dList))
    svList.append(aCost - adList[i])
  
  del adList[0]
  dList[period - 1] = svList[period -1]
  svList[period -1] = 0

  resultList = [dList, adList, svList]

  df1 = pd.DataFrame(resultList, columns = yearCount  , index = ["DE", "AD", "SV"])

  
  print("'DE' = Depreciation amount of each year" + "\n" + "'AD' = Accumulated depreciation amount" + "\n" + "'SV' = Salvage Value")
  print("\n" + "~Result~")
  print("1)On Jan 1st", bYear, ": AD = 0", "SV =", aCost, "\n")
  print("2)The end of term of each year")
  print(df1.T)

if x == 1:
  SLmethod(bYear, eYear, aCost, sValue, period)
elif x == 2:
  SYDmethod(bYear, eYear, aCost, sValue, period)
else:
  DDBmethod(bYear, eYear, aCost, sValue, period)

print("\n")
fin = input("Fin, just ignore this box")
```
